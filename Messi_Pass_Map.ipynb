```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("messibetis.csv")
```

```python
#to convert the data to match the pitch
df["x"]=df["x"]*1.2 
df["y"]=df["y"]*.8
df["endX"]=df["endX"]*1.2
df["endY"]=df["endY"]*.8

```

```python
df
```

```python
fig,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor("#22312b")
ax.patch.set_facecolor("#22312b")

pitch = Pitch(pitch_type='statsbomb', orientation = "horizontal",
              pitch_color='#22312b', line_color='#c7d5cc',figsize=(16, 11),
              constrained_layout=True, tight_layout=False)
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

#for loop to plot every pass, both successful and unsuccessful (differentiated by color)
for x in range(len(df["x"])):
  if df["outcome"][x] == "Successful":
      plt.plot((df["x"][x],df["endX"][x]),(df["y"][x],df["endY"][x]),color="tan")
      plt.scatter(df["x"][x],df["y"][x],color= "tan")
  if df["outcome"][x] == "Unsuccessful":
      plt.plot((df["x"][x],df["endX"][x]),(df["y"][x],df["endY"][x]),color="red")
      plt.scatter(df["x"][x],df["y"][x],color= "red")

#title of visualization
ax.text(x=0.5, y=1.1, s='Messi Pass Map vs Real Betis', color = "white", family = "monospace", fontsize=20, weight='bold', ha='center', va='bottom', transform=ax.transAxes)
ax.text(x=0.5, y=1.05, s='(November 7th, 2020)', color = "white", weight= "bold", fontsize=12, alpha=0.75, ha='center', va='bottom', transform=ax.transAxes)


```
